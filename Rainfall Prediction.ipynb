```python
data.drop('Date', axis = 1, inplace = True)
data
```

```python
categorical_features = [column_name for column_name in data.columns if data[column_name].dtype == 'O']
data[categorical_features].isnull().sum()
```

```python
# Imputing missing values
categorical_features_with_null = [feature for feature in categorical_features if data[feature].isnull().sum()]
for each_feature in categorical_features_with_null:
    mode_val = data[each_feature].mode()[0]
    data[each_feature].fillna(mode_val,inplace=True)
```

```python
# Numerical features
numerical_features = [column_name for column_name in data.columns if data[column_name].dtype != 'O']
data[numerical_features].isnull().sum()
```

```python
# Finding and removing outliers from numerical features
features_with_outliers = ['MinTemp', 'MaxTemp', 'Rainfall', 'Evaporation', 'WindGustSpeed','WindSpeed9am', 'WindSpeed3pm', 'Humidity9am', 'Pressure9am', 'Pressure3pm', 'Temp9am', 'Temp3pm']
for feature in features_with_outliers:
    q1 = data[feature].quantile(0.25)
    q3 = data[feature].quantile(0.75)
    IQR = q3-q1
    lower_limit = q1 - (IQR*1.5)
    upper_limit = q3 + (IQR*1.5)
    data.loc[data[feature]<lower_limit,feature] = lower_limit
    data.loc[data[feature]>upper_limit,feature] = upper_limit
```

```python
# Imputing the missing values
numerical_features_with_null = [feature for feature in numerical_features if data[feature].isnull().sum()]
for feature in numerical_features_with_null:
    mean_value = data[feature].mean()
    data[feature].fillna(mean_value,inplace=True)
```

```python
# EDA ( Exploratory data analysis)
data['RainTomorrow'].value_counts().plot(kind='bar')
```

```python
# Encoding Categorical features
def encode_data(feature_name):
    mapping_dict = {}
    unique_values = list(data[feature_name].unique())
    for idx in range(len(unique_values)):
        mapping_dict[unique_values[idx]] = idx
    return mapping_dict
```

```python
data['RainToday'].replace({'No':0, 'Yes': 1}, inplace = True)

data['RainTomorrow'].replace({'No':0, 'Yes': 1}, inplace = True)

data['WindGustDir'].replace(encode_data('WindGustDir'),inplace = True)

data['WindDir9am'].replace(encode_data('WindDir9am'),inplace = True)

data['WindDir3pm'].replace(encode_data('WindDir3pm'),inplace = True)

data['Location'].replace(encode_data('Location'), inplace = True)
```

```python
# Splitting data as Independent and dependent features
X = data.drop(['RainTomorrow'],axis=1)
y = data['RainTomorrow']
```

```python
# Features Importance
from sklearn.ensemble import ExtraTreesRegressor
etr_model = ExtraTreesRegressor()
etr_model.fit(X,y)
etr_model.feature_importances_
```

```python
feature_imp = pd.Series(etr_model.feature_importances_,index=X.columns)
feature_imp.nlargest(10).plot(kind='barh')
```

```python
# Data into training and testing set
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size = 0.2, random_state = 0)
```

```python
# Checking length of training and testing data
print("Length of Training Data: {}".format(len(X_train)))
print("Length of Testing Data: {}".format(len(X_test)))
```

```python
# Features Scaling
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
```

```python
X_test = scaler.transform(X_test)
```

```python
# Training the model
from sklearn.linear_model import LogisticRegression
classifier_lr = LogisticRegression(solver='liblinear', random_state=0)
classifier_lr.fit(X_train, y_train)
```

```python
# Tesing of model
y_pred = classifier_lr.predict(X_test)
y_pred
```

```python
# Evaluate the model performance 
from sklearn.metrics import accuracy_score
print("Accuracy Score: {}".format(accuracy_score(y_test,y_pred)))
```

```python
# Checking for underfitting and overfitting
print("Train Data Score: {}".format(classifier_lr.score(X_train, y_train)))
print("Test Data Score: {}".format(classifier_lr.score(X_test, y_test)))
```

```python
from sklearn.metrics import confusion_matrix
```

```python
# Checking for confusion matrix
print(confusion_matrix(y_test,y_pred))
```

```python
from sklearn.metrics import classification_report
```

```python
# Checking for classification report
print(classification_report(y_test, y_pred))
```

```python
# Cross validation
from sklearn.model_selection import cross_val_score
scores = cross_val_score(classifier_lr, X_train, y_train, cv = 5, scoring='accuracy')
print('Cross-validation scores:{}'.format(scores))
print('Average cross-validation score: {}'.format(scores.mean()))
```
